# Question 02: [ Marks 10 ]

What is a decision boundary in a perceptron?
Explain how the equation
w₁x₁ + w₂x₂ + b = 0
represents a decision boundary geometrically.

## Write Answer 02:

A decision boundary in a perceptron is a line (in 2D) that separates two classes, such as Class 0 and Class 1. The equation w₁x₁ + w₂x₂ + b = 0 represents this boundary geometrically as a straight line on a graph, where w₁ and w₂ control the direction (slope) of the line and b shifts it up, down, left, or right. Points on one side of the line are classified as 1, and points on the other side are classified as 0. For example, if x₁ is study hours and x₂ is attendance, this line divides students into pass and fail groups.

# Question 03: [ Marks 15 ]

Why is the perceptron called a linear classifier?
Explain your answer using the concept of linear separability and logical gate examples such as AND, OR, and XOR.

## Write Answer 03:
A perceptron is called a linear classifier because it separates data using a straight line (linear decision boundary). This means it can correctly classify data only if the classes are linearly separable (they can be divided by one straight line). Logical gates like AND and OR are linearly separable, so a perceptron can solve them. However, XOR is not linearly separable, meaning no single straight line can separate its outputs  so a single perceptron cannot solve XOR.

## Write Answer 07:

Limitations of a Single-Layer Perceptron

A single-layer perceptron has several limitations. It can only solve linearly separable problems, so it fails on non-linear patterns like XOR. Its learning capacity is limited, as it cannot model complex relationships between inputs and outputs. With only a single decision boundary, it can separate data with a straight line (or hyperplane) but cannot handle more intricate patterns. It also struggles to capture interactions between features, and adding more inputs does not increase its ability to solve non-linear problems.

How Multilayer Perceptrons (MLPs) Help

Using multiple layers in a perceptron, known as a multilayer perceptron (MLP), overcomes these limitations. Hidden layers allow the network to learn non-linear relationships and combine multiple decision boundaries to handle complex patterns. MLPs have greater flexibility and capacity, enabling them to approximate almost any function and capture interactions between features through hidden neurons. This makes them much more effective for real-world problems like image recognition, speech processing, and natural language tasks.
